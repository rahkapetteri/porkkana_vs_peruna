```python
import json
with open('search_index.json') as f:
    data = json.load(f)
```

```python
important_stats = ["hitaus", "hyökkäysvoima", "kestopisteet", "puolustusvoima", "superteho"]
```

```python
for index, competitor in enumerate(data['docs'][1:]):
    for stat in important_stats:
        fullstring = data['docs'][1:][index]['text']
        before_keyword, keyword, after_keyword = fullstring.partition(stat)
        value = after_keyword.strip().partition(' ')[0]
        try:
            data['docs'][1:][index+1][stat] = value
        except IndexError:
            print("Overindex")
```

```python
filteredJson = {}
for index, competitor in enumerate(data['docs'][1:]):
    if data['docs'][1:][index]['text'] == '':
        filteredJson[competitor['title']] = competitor
    else:
        pass
        
```

```python
with open('search_index_parsed.json', "w") as d:
    json.dump(filteredJson, d)
```
