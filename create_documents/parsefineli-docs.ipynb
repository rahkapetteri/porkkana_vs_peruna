```python
def returnKCAL(x):
    if x["Yksikkö"] == "KJ":
        return float(str(x["Arvo"]).replace(",", ".")) / 4.184
#     df["Arvo"] = df["Arvo"].map(lambda x: (float(str(x).replace(",", "."))) / 4.184)
    else:
        return float(str(x["Arvo"]).replace(",", "."))
df["Arvo"] = df.apply(lambda row: returnKCAL(row), axis=1)
df["Yksikkö"].replace("KJ", "kcal", inplace=True)
```

```python
def returnUNIT(x):
    return str(round(float(x["Arvo"]),2)) + " " + x["Yksikkö"]

df["VALS"] = df["Arvo"].copy()
df["Arvo"] = df.apply(lambda row: returnUNIT(row), axis=1)
```

```python
def convertMG(x):
    if x["Yksikkö"] == "mg":
        x["VALS"] = x["VALS"] / 1000
        return round(x["VALS"], 2)
    elif x["Yksikkö"] == "μg":
        x["VALS"] = x["VALS"] / 1000000
        return round(x["VALS"], 2)
    else:
        return round(x["VALS"], 2)
df["VALS"] = df.apply(lambda row: convertMG(row), axis=1)
```

# Statsit:

> statsit täytyy löytyä ja olla kaavan mukaan. On kuitenkin lupa lisätä taisteluihin satunnaisuutta.
100 grammaa sisältää:

Energia (kcal) = Health Points eli kestopisteet.
Hiilihydraatit (g) = Hyökkäysvoima
Proteiinit (g) = Puolustusvoima (voidaan käyttää esim. prosentuaalisesti, koska maksimi on tietty 100)
Rasvat ei erikseen lisää mitään statsia, mutta enemmän rasvaa = enemmän energiaa = enemmän helaa
Hiilihydraattien, rasvojen ja proteiinien yhteenlaskettu grammamäärä = Hitaus. (tai käänteisellä arvolla hyökkäysnopeus)
