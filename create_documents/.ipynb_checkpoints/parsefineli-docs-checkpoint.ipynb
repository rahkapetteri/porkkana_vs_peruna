```python
def returnKCAL(x):
    if x["Yksikkö"] == "KJ":
        return float(str(x["Arvo"]).replace(",", ".")) / 4.184
#     df["Arvo"] = df["Arvo"].map(lambda x: (float(str(x).replace(",", "."))) / 4.184)
    else:
        return float(str(x["Arvo"]).replace(",", "."))
df["Arvo"] = df.apply(lambda row: returnKCAL(row), axis=1)
df["Yksikkö"].replace("KJ", "kcal", inplace=True)
```

```python
def returnUNIT(x):
    return str(round(float(x["Arvo"]),2)) + " " + x["Yksikkö"]

df["VALS"] = df["Arvo"].copy()
df["Arvo"] = df.apply(lambda row: returnUNIT(row), axis=1)
```

```python
def convertMG(x):
    if x["Yksikkö"] == "mg":
        x["VALS"] = x["VALS"] / 1000
        return round(x["VALS"], 2)
    elif x["Yksikkö"] == "μg":
        x["VALS"] = x["VALS"] / 1000000
        return round(x["VALS"], 2)
    else:
        return round(x["VALS"], 2)
df["VALS"] = df.apply(lambda row: convertMG(row), axis=1)
```

```python
df.Kategoria.replace(", rasvaa <= 1 %", " vähärasvainen", inplace=True, regex=True)
df.Kategoria.replace(", rasvaa > 1 %", "", inplace=True, regex=True)
df.Kategoria.replace("rasvaa  <= 17%", "vähärasvainen", inplace=True, regex=True)
df.Kategoria.replace(", rasvaa >17 %", "", inplace=True, regex=True)
df.Kategoria.replace("rasvaa <= 17 %", "vähärasvainen", inplace=True, regex=True)
df.Kategoria.replace(", rasvaa > 17 %", "", inplace=True, regex=True)
df.Kategoria.replace("rasvaa <= 15 %", "vähärasvainen", inplace=True, regex=True)
df.Kategoria.replace(", rasvaa >15 %", "", inplace=True, regex=True)
df.Kategoria.replace(" <55%"," alle 55 pros", inplace=True, regex=True)
df.Kategoria.replace(" >= 55%", "", inplace=True, regex=True)
df.Kategoria.replace(", rasvaa > 2 % ", "", inplace=True, regex=True)
df.Kategoria.replace(" < 55 %"," alle 55 pros", inplace=True, regex=True)
df.Kategoria.replace(">= 55 %", "", inplace=True, regex=True)
df.Kategoria.replace(", rasvaa <= 2 %", "", inplace=True, regex=True)
df.Kategoria.replace("Kerma/ creme", "Kerma" ,inplace=True, regex=True)
df.Kategoria.replace("  ", " ")
df.Kategoria.replace("Juusto, kasvirasva-", "Juusto, kasvirasva", inplace=True, regex=True)
df.Kategoria.replace("(sis. mehiläisvalmiste)", "sis mehiläisvalmiste", inplace=True, regex=True)
df.Kategoria = df.Kategoria.map(lambda x: x.strip())
```

```python
df["filename"] = df["Raaka-aine"].map(lambda x: re.sub(r'[^\w\d_]','-',x).lower().replace("ä", "a").replace("ö", "o"))
df.filename = df.filename.map(lambda x: x.replace("-----","-").replace("----","-").replace("---", "-").replace("--","-"))
df.filename = df.filename.map(lambda x: x[:-1] if x[-1] == "-" else x)
df.filename = df.filename.map(lambda x: x[1:] if x[0] == "-" else x)
df["Raaka-aine"] = df["Raaka-aine"].map(lambda x: x.replace("  ", " ").replace("   ", " "))
```
